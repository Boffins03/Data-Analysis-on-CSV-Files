```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load CSV
df = pd.read_csv("sales_data.csv")

# Show first 5 rows
print("First 5 rows of the dataset:")
display(df.head())

# Basic info
print("\nDataset Info:")
df.info()

print("\nSummary Statistics:")
display(df.describe())

print("\nNull Values:")
display(df.isnull().sum())
```

```python
# Group by product and sum sales
sales_by_product = df.groupby("Product")["Sales"].sum().reset_index()
print("\nTotal sales by product:")
display(sales_by_product)
```

```python
# Group by month (if 'Date' column exists)
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df["Month"] = df["Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Sales"].sum().reset_index()
print("\nTotal sales by month:")
display(monthly_sales)

# Plot monthly sales
monthly_sales.plot(x="Month", y="Sales", kind="bar", figsize=(8,5))
plt.title("Monthly Sales")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()

# 5. Plot sales by product
sales_by_product.plot(x="Product", y="Sales", kind="bar", figsize=(8,5))
plt.title("Sales by Product")
plt.ylabel("Total Sales")
plt.xlabel("Product")
plt.show()
```
